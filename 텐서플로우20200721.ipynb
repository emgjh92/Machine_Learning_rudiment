```
#https://raw.githubusercontent.com/hunkim/DeepLearningZeroToAll/master/data-03-diabetes.csv

# Lab 5 Logistic Regression Classifier
import tensorflow as tf
import numpy as np

tf.set_random_seed(777)  # for reproducibility

xy = np.loadtxt('https://raw.githubusercontent.com/hunkim/DeepLearningZeroToAll/master/data-03-diabetes.csv', delimiter=',', dtype=np.float32)
#2중 리스트 형태로 배열에 데이터를 담는다.
#ex) print(xy[0][0])
#ex) print(xy[0][8])
#ex) print(xy)

x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape) #.shape = 몇행 몇열로 이루어져 있냐
print(y_data.shape)

# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 8])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([8, 1]), name='weight')#weight => 입력계수, 출력계수
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis using sigmoid: tf.div(1., 1. + tf.exp(-tf.matmul(X, W)))
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)

# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) *
                       tf.log(1 - hypothesis))

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
#GradientDescent 알고리즘

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

# Launch graph
with tf.Session() as sess:
    
    sess.run(tf.global_variables_initializer())

    for step in range(100000) :
      sess.run(train, feed_dict={X: x_data, Y: y_data})

      if step%200 == 0:
            tc = sess.run(cost,feed_dict={X: x_data, Y: y_data})
            print("%10d %10.7f" %(step,tc))
      ta = sess.run(accuracy,feed_dict={X: x_data, Y: y_data})
      print("예측 정확도 : ", ta) #퍼센테이지 임
      #예측 정확도는 range 를 많이 돌릴 수록 높아 짐
      x_test = [
                [7.,7.,7.,7.,7.,7.,7.,7.]
      ]
      tp = sess.run(predicted, feed_dict={X: x_test})
      print("에측 결과 : ", tp)
```
